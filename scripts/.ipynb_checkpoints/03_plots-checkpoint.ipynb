```python
europe.loc[("Italy", "Ireland", "Iceland"), ] # we want only Italy, Ireland and Iceland as rows; we want all the columns so we put nothing after the comma
```

```python
gdp_i = europe.loc[("Italy", "Ireland", "Iceland"), ]
```

```python
plt.plot(years, gdp_i.iloc[0], 'g--', label='Italy')
plt.plot(years, gdp_i.iloc[1], 'y--', label='Ireland')
plt.plot(years, gdp_i.iloc[2], 'r--', label='Iceland')
# g-- is green; y-- is yellow; r-- is red
plt.legend(loc = 'best')
# add a legend, set location to 'best' and it will find the best location
plt.xlabel('Years')
plt.ylabel('GDP per Cap')
plt.title('GDP per Cap in Italy, Ireland and Iceland')
plt.savefig('../plots/GDP_i_countries.png')
# save plot as a figure in a location of choice; you NEED to add an extension
```
